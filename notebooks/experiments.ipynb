```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load the Boston Housing Dataset

```python
boston_df = pd.read_csv("HousingData.csv")
boston_df.head()
```

```python
boston_df.info()
```

```python
boston_df.describe()
```

## Prepare the Dataframe

```python
boston_df.rename(columns={"CRIM":"criminal",
      "ZN":"land_zones",
      "INDUS":"small_business",
      "CHAS":"river_neighboor",
      "NOX":"nitric_oxides",
      "RM":"avg_room",
      "AGE":"age",
      "DIS":"distance_center",
      "RAD":"highway_access",
      "TAX":"tax",
      "PTRATIO":"student_teacher_ratio",
      "B":"black_people",
      "LSTAT":"low_population",
      "MEDV":"price"}, inplace=True)
boston_df.head()
```

```python
# Check for null values
boston_df.isnull().sum()
```

```python
# Fill null values with 0
boston_df.fillna(0, inplace=True)
boston_df.isnull().sum()
```

```python
# EDA

boston_df.corr()
```

```python
sns.pairplot(boston_df)
```

```python
sns.regplot(x='avg_room', y='price', data=boston_df)
```

```python
sns.regplot(x='low_population', y='price', data=boston_df)
```

```python
# Prepare data 
X = boston_df.iloc[:, :-1]
y = boston_df.iloc[:, -1]
print(X.head())
print(y.head())
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# print coefficients and intercept
print("Coefficients: ", lr.coef_)
print("Intercept: ", lr.intercept_)
```

```python
# Predictions
y_pred = lr.predict(X_test)     
y_pred
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
```

```python
# Plot the Residuals
residuals = y_test - y_pred
sns.displot(residuals, kind='kde')
```

```python
# Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# Adjusted R2 Score
adjusted_r2 = 1 - (1 - r2_score(y_test, y_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
print("Adjusted R2 Score:", adjusted_r2)
```

## New Data Prediction

```python
lr.predict(scaler.transform(X_test[0, :].reshape(1,-1)))
```
